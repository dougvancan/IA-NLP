```
#Treine o modelo de classificacao Decision Tree para classificar os produtos em suas categorias

import pandas as pd

# Import do data frame
url = 'https://s3.amazonaws.com/automl-example/produtos.csv'

# pandas read_csv with skiprows option
df = pd.read_csv(url, sep = ';', encoding='utf-8')
df.describe()

#Crie uma nova coluna com a juncao do nome e descricao dos produtos
df['text'] = df['nome'] + " " +df['descricao']


# Remover linhas com valores nulos
df.dropna(inplace=True)
df.shape
#df.nome_descricao[0]
df.head()
etejo de
```

```
#Contagem de Termos - UNIGRAMA
from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(ngram_range=(1,1))
vect.fit(df.text)
count_vect = vect.transform(df.text)
df.text.shape

```

```
#Treinamento
# Modelo n grama
from sklearn.tree import DecisionTreeClassifier


tree = DecisionTreeClassifier()
tree.fit(count_vect, df['categoria'])

print('D Tree: ', tree.score(count_vect, df['categoria']))
print(tree.classes_)
```

```
#Teste diferentes configuracoes de n-grama e vetorizacao
texto = vect.transform(['livro do harry'])

print('D Tree: ', tree.predict(texto))
```
