```
import pandas as pd

# Import do data frame
url = 'https://s3.amazonaws.com/automl-example/produtos.csv'

# pandas read_csv with skiprows option
df = pd.read_csv(url, sep = ';', encoding='utf-8')
df.describe()
```

```
#Crie uma nova coluna com a juncao do nome e descricao dos produtos
df['nome_descricao'] = df['nome'] + " " +df['descricao']


# Remover linhas com valores nulos
df.dropna(inplace=True)
df.shape
#df.nome_descricao[0]
```

```
#Quantos unigramas bigramas, trigramas existem nessa nova coluna antes e depois de remover as stopwords?


import nltk
nltk.download('stopwords')


from sklearn.feature_extraction.text import CountVectorizer

vect = CountVectorizer(ngram_range=(1,1))
vect.fit(df.nome_descricao)
text_vect = vect.transform(df.nome_descricao)

print("Shape ", text_vect.shape)
print("Unigramas com StopWords", text_vect.shape[1])

#print(pd.DataFrame(text_vect.A, columns=vect.get_feature_names()).to_string())
```
